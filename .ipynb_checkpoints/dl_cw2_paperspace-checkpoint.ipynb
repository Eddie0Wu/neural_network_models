```python
# *CODE FOR PART 1.1a IN THIS CELL*

class VAE(nn.Module):
    def __init__(self, latent_dim):
        super(VAE, self).__init__()
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################

        # self.encoder = nn.Sequential(
        #     nn.Conv2d(1, 16, kernel_size = 3),
        #     nn.BatchNorm2d(16),
        #     nn.LeakyReLU(),
        #     nn.Conv2d(16, 32, kernel_size = 3),
        #     nn.BatchNorm2d(32),
        #     nn.LeakyReLU(),
        #     nn.Conv2d(32, 64, kernel_size = 3, stride = 2),
        #     nn.BatchNorm2d(64),
        #     nn.LeakyReLU(),
        #     nn.Conv2d(64, 128, kernel_size = 3, stride = 2),
        #     nn.BatchNorm2d(128),
        #     nn.LeakyReLU()
        # )

        # self.fc_mean = nn.Linear(128*5*5, latent_dim)
        # self.fc_logvar = nn.Linear(128*5*5, latent_dim)
        # self.fc_grow = nn.Linear(latent_dim, 128*5*5)

        # self.decoder = nn.Sequential(
        #     nn.ConvTranspose2d(128, 64, kernel_size = 3, stride = 2),
        #     nn.BatchNorm2d(64),
        #     nn.LeakyReLU(),
        #     nn.ConvTranspose2d(64, 32, kernel_size = 4, stride = 2),
        #     nn.BatchNorm2d(32),
        #     nn.LeakyReLU(),
        #     nn.ConvTranspose2d(32, 16, kernel_size = 3),
        #     nn.BatchNorm2d(16),
        #     nn.LeakyReLU(),
        #     nn.ConvTranspose2d(16, 1, kernel_size = 3),
        #     nn.Sigmoid()
        # )
        ### I tried implementing convolutional VAE too, but found 
        ### fully-connected VAE to perform better, hence I chose to 
        ### stick with fully-connected VAE and commented out the codes
        ### for convolutional VAE.

        self.latent_dim = latent_dim

        self.encoder = nn.Sequential(
            nn.Linear(28*28, hidden_layer),
            nn.LeakyReLU(),
            nn.Linear(hidden_layer, 100),
            nn.LeakyReLU(),
        )

        self.fc_mean = nn.Linear(100, latent_dim)
        self.fc_logvar = nn.Linear(100, latent_dim)

        self.decoder = nn.Sequential(
            nn.Linear(self.latent_dim, 100),
            nn.LeakyReLU(),
            nn.Linear(100, hidden_layer),
            nn.LeakyReLU(),
            nn.Linear(hidden_layer, 28*28),
            nn.Sigmoid()
        )

        #######################################################################
        #                       ** END OF YOUR CODE **
        ####################################################################### 
        
    def encode(self, x):
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################

        # x = self.encoder(x)
        # x = x.view(x.shape[0], -1)
        # mu = self.fc_mean(x)
        # logvar = self.fc_logvar(x)

        x = x.view(x.shape[0], -1)
        x = self.encoder(x)
        mu = self.fc_mean(x)
        logvar = self.fc_logvar(x)

        return mu, logvar

        #######################################################################
        #                       ** END OF YOUR CODE **
        ####################################################################### 
    
    def reparametrize(self, mu, logvar):
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################

        std = torch.exp(0.5*logvar)
        eps = torch.randn_like(std)
        return mu + eps*std

        #######################################################################
        #                       ** END OF YOUR CODE **
        ####################################################################### 

    def decode(self, z):
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################


        z = self.decoder(z)
        z = z.view(z.shape[0], 1, 28, 28)
        return z


        #######################################################################
        #                       ** END OF YOUR CODE **
        ####################################################################### 
    
    def forward(self, x):
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################

        mu, logvar = self.encode(x)
        z = self.reparametrize(mu, logvar)
        out = self.decode(z)
        return out, mu, logvar
        #######################################################################
        #                       ** END OF YOUR CODE **
        ####################################################################### 
```

```python
# VAE model 
model = VAE(latent_dim).to(device)
params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print("Total number of parameters is: {}".format(params))
print(model)
# optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

## Part 1.1b: Training the Model (5 Points)

### Defining a Loss
Recall the Beta VAE loss, with an encoder $q$ and decoder $p$:
$$ \mathcal{L}=\mathbb{E}_{q_\phi(z \mid X)}[\log p_\theta(X \mid z)]-\beta D_{K L}[q_\phi(z \mid X) \| p_\theta(z)]$$

In order to implement this loss you will need to think carefully about your model's outputs and the choice of prior.

There are multiple accepted solutions. Explain your design choices based on the assumptions you make regarding the distribution of your data.

* Hint: this refers to the log likelihood as mentioned in the tutorial. Make sure these assumptions reflect on the values of your input data, i.e. depending on your choice you might need to do a simple preprocessing step.

* You are encouraged to experiment with the weighting coefficient $\beta$ and observe how it affects your training

```python
# *CODE FOR PART 1.1b IN THIS CELL*

def loss_function_VAE(recon_x, x, mu, logvar, beta = 1):
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################
        
        # mse = nn.MSELoss()
        # recon_loss = F.mse_loss(recon_x, x)
        # kl_divergence = torch.mean(-0.5*torch.sum(1 + logvar - mu**2 - logvar.exp(), dim = 1), dim = 0)
        # loss = recon_loss + beta*kl_divergence

        bce = F.binary_cross_entropy(recon_x, x, reduction = 'sum')
        kl_divergence = -0.5*torch.sum(1 + logvar - mu**2 - logvar.exp())
        loss = bce + beta*kl_divergence
        return loss, bce, kl_divergence


        #######################################################################
        #                       ** END OF YOUR CODE **
        ####################################################################### 

print_every = 200

# I wrote a function for testing
def test_part(loader, model, beta = 1):
        # record the various losses
        recon_loss = 0
        kl_div = 0
        total_loss = 0
        count = 0
        model.eval()
        with torch.no_grad():
          for x,y in loader:
            x = x.to(device = device)
            y = y.to(device = device)
            # pass input through the model
            recon_x, mu, logvar = model(x)
            loss, rec_loss, kld = loss_function_VAE(recon_x, x, mu, logvar, beta = beta)
            # update the losses
            recon_loss += rec_loss
            kl_div += kld
            total_loss += loss
            count += 1
          print(f'test loss: {total_loss/count:.5f}')
          return (total_loss/count), (recon_loss/count), (kl_div/count)
```

```python
# model training loop

# for plotting
train_losses = []
train_recon_losses = []
train_klds = []
test_losses = []
test_recon_losses = []
test_klds = []

for epoch in range(num_epochs):     
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################

        for t, (data, _) in enumerate(loader_train):
          model.train()
          data = data.to(device = device)
          # train the model
          recon_x, mu, logvar = model(data)
          loss, recon_loss, kld = loss_function_VAE(recon_x, data, mu, logvar, beta = 2)
          optimizer.zero_grad()
          loss.backward()
          optimizer.step()

          if t%print_every == 0:
            print(f'Epoch:{epoch}, Iteration:{t}, Loss = {loss.item():.5f}')
            train_losses.append(loss)
            train_recon_losses.append(recon_loss)
            train_klds.append(kld)
            aa,bb,cc  = test_part(loader_test, model, beta = 2)
            test_losses.append(aa)
            test_recon_losses.append(bb)
            test_klds.append(cc)

        #######################################################################
        #                       ** END OF YOUR CODE **
        ####################################################################### 
        
        # save the model
        if epoch == num_epochs - 1:
            with torch.no_grad():
                torch.jit.save(torch.jit.trace(model, (data), check_trace=False),
                    '/content/drive/MyDrive/icl_dl_cw2/CW_VAE/VAE_model.pth')
```

### Loss Explanation
Explain your choice of loss and how this relates to:

* The VAE Prior
* The output data domain
* Disentanglement in the latent space

**YOUR ANSWER**

My loss function is "reconstruction loss + KL regulariser". I chose the VAE prior, Z, to be multivariate standard Gaussian distribution, with covariance matrix equal to identity matrix. Data is assumed to be generated conditioning on the latent variable, Z, from p( x | Z).

To ensure the regularity of the latent space and to prevent excessive overfitting, q( Z | x) is designed to be the factorised multivariate Gaussian distribution parameterised by the neural network. This means that different components of Z have zero covariance with one another, hence helping to achieve regularity and disentanglement in the latent space.

Since both q(Z |x) and p(Z) are factorised Gaussian distribution, the KL divergence term can have an analytic form which I have used as the KL regulariser in my loss function (refer to the previous cell with loss function). 

For the reconstruction loss, I notice that the MNIST image pixels in this exercise are (categorically) distributed in the range of [0, 1]. I use sigmoid activation at the end of the decoder, and binary cross entropy (which measures how different the generated output is from the actual image) for calculating the reconstruction loss. Although BCE is not theoretically accurate as a loss function in this case (since MNIST pixels do not have a binary distribution), it provides fairly good empirical results.

To sum up, the loss function tries to minimise reconstruction loss by making the generated output as similar to the actual image as possible. At the same time, this is regularised by the KL term which measures how far q( Z | x) is from my prior P(Z). This constrains the model to produce a regularised latent space eventually.

**YOUR ANSWER**

From the previous curve plots, I observe that log_likelihood loss (e.g. reconstruction loss) decreases while KL loss increases during training. Both losses change very drastically in the initial epochs and gradually plateau as training continues. The extent of increase in KL loss is outweighed by the extent of fall in log_likehood loss, resulting in the fall of overall loss.

I believe this behaviour is desirable because it is in line with my expectation given the loss function design. In the early epochs, as the network tries to reduce reconstruction loss, q(Z | x) gradually moves away from the assumed prior distribution p(Z), hence causing KL loss to rise. This implies that the network is trying to reconstruct images as close to the actual images as possible, while being penalised by the KL divergence to prevent overfitting to training data.

I do not observe posterior collapse during training. Although KL loss is very small in the beginning, it increases very quickly in the first two epochs and has not collapsed since. 

Posterior collapse happens when training ends up in a degenerate local minimum where the model learns to ignore the latent encoding and the posterior mimics the prior. One possible explanation for why I do not observe posterior collapse might be that my fully-connected VAE is not sophisticated enough to reduce reconstruction loss while ignoring the latent encoding. This might have forced the VAE to use the latent encoding to reconstruct better images. Also, using the appropriate hyperparameters such as learning rate might have helped avoid this.

From the earlier visualisation, I observe that the reconstructed images are extremely similar to the actual images, other than being slightly blurry which is a common shortcoming of VAE. With regard to the randomly generated images, as I have chosen a centred unit variance Gaussian prior, most of the generated images from the random Z contain rather comprehensible digits.

When I experimented with other beta values ranging from 0.1 to 20, I notice from the loss curves that the larger the beta, the smaller the KL loss and hence the wider the spread between reconstruction loss and KL loss (please refer to my submitted "final_notebook" zip for the loss plots with various beta values). Experimenting with different beta value reveals the evident trade-off between reconstruction loss and KL loss. For instance, a larger beta implies that KL loss has a larger weight in the loss function, and the model will force KL loss to increase less. This means that the VAE's learned latent space deviates less from the prior and is less capable of reconstructing the images, hence the reconstruction loss does not fall too much, resulting in the large spread between recontruction loss and KL loss (refer to my plots). 

On the other hand, when beta is small, e.g. 0.1, KL loss has a smaller weight in the loss function and the VAE can better fit the latent space to the data. This leads to higher KL loss and lower reconstruction loss, hence narrowing the spread between them (refer to my plots). However, this is not necessarily desirable as the latent space might have overfitted to the data and have less regularity. Looking at the randomly generated images from part 1.2b, I notice that when beta = 0.1, the generated images are gibberish, whereas when beta = 2, the generated images are mostly recognisable digits. This implies that with a very small beta, although reconstruction effects are better, the VAE is less able of generalisation in the latent space.

Therefore, beta is a hyperparameter whose optimal value needs to be determined via tuning. The value of beta determines how much regularisation is injected in the VAE. In my submitted VAE model, after weighing the trade-off between the reconstruction image quality and the randomly generated image quality, I decided to use beta = 2 because it offers a fairly regularised latent space as well as decent reconstruction quality.

#### Model Definition
Define hyperparameters and the model

```python
# *CODE FOR PART 2.1 IN THIS CELL*

# Choose the number of epochs, the learning rate
# and the size of the Generator's input noise vetor.

num_epochs = 25
learning_rate = 0.0002
latent_vector_size = 100

# Other hyperparams
num_feature_maps = 128
```

```python
# *CODE FOR PART 2.1 IN THIS CELL*


class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################

        self.gen = nn.Sequential(
            nn.ConvTranspose2d(latent_vector_size, num_feature_maps*4, 4, 1, 0, bias = False),
            nn.BatchNorm2d(num_feature_maps*4),
            nn.ReLU(),
            nn.ConvTranspose2d(num_feature_maps*4, num_feature_maps*2, 4, 2, 1, bias = False),
            nn.BatchNorm2d(num_feature_maps*2),
            nn.ReLU(),
            nn.ConvTranspose2d(num_feature_maps*2, num_feature_maps, 4, 2, 1, bias = False),
            nn.BatchNorm2d(num_feature_maps),
            nn.ReLU(),
            nn.ConvTranspose2d(num_feature_maps, 3, 4, 2, 1, bias = False),
            nn.Tanh()
        )

        #######################################################################
        #                       ** END OF YOUR CODE **
        ####################################################################### 


    def forward(self, z, label = None):
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################
        
        out = self.gen(z)

        #######################################################################
        #                       ** END OF YOUR CODE **
        #######################################################################
        return out


class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################
        
        self.dis = nn.Sequential(
            nn.Conv2d(3, num_feature_maps, 4, 2, 1, bias = False),
            nn.BatchNorm2d(num_feature_maps),
            nn.LeakyReLU(0.2),
            nn.Conv2d(num_feature_maps, num_feature_maps*2, 4, 2, 1, bias = False),
            nn.BatchNorm2d(num_feature_maps*2),
            nn.LeakyReLU(0.2),
            nn.Conv2d(num_feature_maps*2, num_feature_maps*4, 4, 2, 1, bias = False),
            nn.BatchNorm2d(num_feature_maps*4),
            nn.LeakyReLU(0.2),
            nn.Conv2d(num_feature_maps*4, 1, 4, 1, 0, bias = False),
            nn.Sigmoid()
        )

        #######################################################################
        #                       ** END OF YOUR CODE **
        ####################################################################### 
        
    def forward(self, x, label = None):
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################
     
        out = self.dis(x)
        #######################################################################
        #                       ** END OF YOUR CODE **
        ####################################################################### 
        
        return out
```

<h2> Initialize Model and print number of parameters </h2>

You can use method `weights_init` to initialize the weights of the Generator and Discriminator networks. Otherwise, implement your own initialization, or do not use at all. You will not be penalized for not using initialization.

```python
# custom weights initialization called on netG and netD
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        m.weight.data.normal_(0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        m.weight.data.normal_(1.0, 0.02)
        m.bias.data.fill_(0)
```

```python
use_weights_init = True

model_G = Generator().to(device)
if use_weights_init:
    model_G.apply(weights_init)
params_G = sum(p.numel() for p in model_G.parameters() if p.requires_grad)
print("Total number of parameters in Generator is: {}".format(params_G))
print(model_G)
print('\n')

model_D = Discriminator().to(device)
if use_weights_init:
    model_D.apply(weights_init)
params_D = sum(p.numel() for p in model_D.parameters() if p.requires_grad)
print("Total number of parameters in Discriminator is: {}".format(params_D))
print(model_D)
print('\n')

print("Total number of parameters is: {}".format(params_G + params_D))
```

### Part 2.1b: Training the Model (12 Points)

#### Defining a Loss

```python
def loss_function(out, label):
    loss = F.binary_cross_entropy(out, label)
    return loss
```

<h3>Choose and initialize optimizers</h3>

```python
# setup optimizer
# You are free to add a scheduler or change the optimizer if you want. We chose one for you for simplicity.
beta1 = 0.5
optimizerD = torch.optim.Adam(model_D.parameters(), lr=learning_rate, betas=(beta1, 0.999))
optimizerG = torch.optim.Adam(model_G.parameters(), lr=learning_rate, betas=(beta1, 0.999))

# Set up learning rate scheduler
schedulerD = StepLR(optimizerD, step_size=6, gamma = 0.5)
schedulerG = StepLR(optimizerG, step_size=6, gamma = 0.5)
```

<h3> Define fixed input vectors to monitor training and mode collapse. </h3>

```python
real_label = 1
fake_label = 0
```

#### Training Loop

```python
# record the losses
train_losses_D = []
train_losses_D_real = []
train_losses_D_fake = []
train_losses_G = []


for epoch in range(num_epochs):

    train_loss_D = 0
    train_loss_D_real = 0
    train_loss_D_fake = 0
    train_loss_G = 0

    for i, data in enumerate(loader_train, 0):
        
        ############################
        # (1) Update D network: maximize log(D(x)) + log(1 - D(G(z)))
        ###########################device
        
        # train with real
        model_D.zero_grad()

        real_cpu = data[0].to(device = device)
            # create the labels
        batch = real_cpu.shape[0]
        label = torch.full((batch,), real_label, dtype = torch.float, device = device)
        out = model_D(real_cpu).view(-1)
            # calculate loss and propagate back
        real_error = loss_function(out, label)
        real_error.backward()

        D_x = out.mean().item()


        # train with fake
        fixed_noise = torch.randn(batch, latent_vector_size, 1, 1, device=device)
            # generate fake samples
        fake = model_G(fixed_noise)
        label.fill_(fake_label)
        out = model_D(fake.detach()).view(-1)

        fake_error = loss_function(out, label)
        fake_error.backward()

        D_G_z1 = out.mean().item()

        errD = real_error + fake_error  

        optimizerD.step()     
            # keep track of the losses for plotting
        train_loss_D += errD.item()
        train_loss_D_real += real_error.item()
        train_loss_D_fake += fake_error.item()


        ############################
        # (2) Update G network: maximize log(D(G(z)))
        ###########################
        model_G.zero_grad()
            # generate labels to be real
        label.fill_(real_label)
        out = model_D(fake).view(-1)

            # calculate loss and propagate back
        errG = loss_function(out, label)
        errG.backward()

        D_G_z2 = out.mean().item()
            # update optimiser
        optimizerG.step()
            # keep track of the loss for plotting
        train_loss_G += errG.item()


        print('[%d/%d][%d/%d] Loss_D: %.4f Loss_G: %.4f D(x): %.4f D(G(z)): %.4f / %.4f'
              % (epoch, num_epochs, i, len(loader_train),
                 errD.item(), errG.item(), D_x, D_G_z1, D_G_z2))

    # step the learning rate scheduler
    schedulerD.step()
    schedulerG.step()

    if epoch == 0:
        save_image(denorm(real_cpu.cpu()).float(), '/content/drive/MyDrive/icl_dl_cw2/CW_GAN/real_samples.png')
    with torch.no_grad():
        fake = model_G(fixed_noise)
        save_image(denorm(fake.cpu()).float(), '/content/drive/MyDrive/icl_dl_cw2/CW_GAN/fake_samples_epoch_%03d.png' % epoch)
    train_losses_D.append(train_loss_D / len(loader_train))
    train_losses_D_real.append(train_loss_D_real / len(loader_train))
    train_losses_D_fake.append(train_loss_D_fake / len(loader_train))
    train_losses_G.append(train_loss_G / len(loader_train))


   
    
# save  models 
# if your discriminator/generator are conditional you'll want to change the inputs here
torch.jit.save(torch.jit.trace(model_G, (fixed_noise)), '/content/drive/MyDrive/icl_dl_cw2/CW_GAN/GAN_G_model.pth')
torch.jit.save(torch.jit.trace(model_D, (fake)), '/content/drive/MyDrive/icl_dl_cw2/CW_GAN/GAN_D_model.pth')

```

* Are there any tricks you came across in the literature etc. which you suspect would be helpful here

**Your Answer**

In the generator, I use 4 transpose convolutional layers (some with stride) to generate 32*32 images. After every transpose convolutional layer, I add batch normalisation (which injects noise during training to help with regularisation, and in the past people believe that batchnorm can also address covariate shift) before passing through ReLU activation. Tanh is used after the final convolution to output values between -1 and 1, because the image data is normalised during preprocessing. In the discriminator, I also use 4 convolutional layers (some with stride) with batch normalisation, before passing through LeakyReLU activation function according to DCGAN literatures (which find LeakyReLU to perform better). Sigmoid is used after the final convolution to output value between 0 and 1 to indicate whether the image is real or fake.

For both networks, I choose kernel size = 4 (according to literatures) and channel size to be a multiple of 128 (after doing hyperparameter search). Convolutional layers have no bias. Weights are initialised with random normal distribution N(0, 0.02) according to the DCGAN paper. For the loss function of the generator network, I maximise log(D(G(z))), the probability of the discriminator classifying fake image as real, so that loss is "non-saturated". I also used a learning rate scheduler which halves the learning rate every 10 epochs.

I first tried to implement several ResNet blocks (replace max pooling with upsampling, and conv2d with convtranspose2d in the generator) as the network architecture but losses do not decrease during training. Then, I simplified the network architecture by having fever convolutions and removing upsampling, and with the reference of online literatures, arrived at the current architecture. After experimenting with DCGAN, the things that did not work are: too many convolutional layers, using upsampling in the generator, small kernel size, batch size greater than 256, learning rate below 0.0001. The things that worked are: having 4 (transpose) convolutional layers, kernel size = 4, using LeakyReLU activation for the discriminator, using batch normalisation, using a smaller batch size, using a learning rate step scheduler, having 128 channels before the final tranpose convolution into 3 channels in the generator

Judging from the loss curves, I find the training process to be rather unstable, and having read through the original DCGAN paper again, I realised that applying batchnorm to all layers leads to sample oscillation and model instability (Radford, Metz & Chintala 2016). Following the original paper, I tried to avoid this by removing batchnorm from the generator output layer and the discriminator input layer. However, this leads to very evident partial mode collapse (might be because my model is simplier than the original paper). After trial and error, I avoided partial mode collapse by adding batchnorm back to the discriminator input layer, and obtained better model performance.

I found that the number of (transpose) convolutional layers, strided convolution, kernel size, learning rate and batch size mattered the most. When there are too many convolutional layers, the training is highly unstable and the generated images are unsatisfactory. This might be because when both networks are too massive, they struggle to reach an equilibrium while competing. When kernel size is smaller or larger than 4, or when the batch size is too large, generated images tend to have many specks, stripes or square boxes. In addition, I notice that given the CIFAR-10 dataset at hand, longer training does not necessarily lead to better result. Hence, I monitor the losses during training (and at the same time observe the generated fake images) and stop when the generator performance reaches a satisfactory level, usually around epoch = 16. In addition, decaying the learning rate over time tends to produce more stable training. I started with lr = 0.0002, and halves it every 10 epochs.

With my current model, a few generated images still contain bright specks. One major challenge of GAN is the inherent instability in training. Some tricks I came across in literatures include: 1. training with labels for the images which can lead to huge improvement in the quality of generated samples; 2. using virtual batch normalisation which calculates normalisation batch statistics with a union of real and fake images (suggested by Ian Goodfellow in his 2016 paper titled "Tutorial: Generative Adversarial Networks").

## Part 2.3: Understanding Mode Collapse (5 points) 
**Your task:** 

Based on the images created by your generator using the `fixed_noise` vector during training, provide a discussion on whether you noticed any mode collapse, what this behaviour may be attributed to, and explain what you did in order to cope with mode collapse.

```python
# Any additional code
```

### Discussion


**YOUR ANSWER**

Mode collapse happens when the generator can only generate a few modes, i.e. a large number of points in the latent space are mapped to a small subset of generated images. 

I do not observe any complete mode collapse in the generated images during training. It is worth mentioning that in the first two epochs, many generated images look similar. This might be due to the fact that the generator was trying to learn the latent representations in the beginning and hence unable to generate very distinct images. As the generator performance improves from epoch 3 onwards, the generated images look much more distinct. Also, during training, I notice that several generated images look similar occasionally and this might indicate partial mode collapse.

Mode collapse happens when the generator is being updated without updates to the discriminator (for example, due to vanishing gradient), generated images might converge to the optimal point that fools the discriminator the most, and the optimal point will be independent of the latent distribution in the extreme, resulting in mode collapse to a single point.

When I spot such partial mode collapse, I stop the training, adjust the hyperparameters such as learning rate and hidden channels, as well as the network architecture (as aforementioned, I found that the removal of batchnorm from discriminator input layer leads to greater partial mode collapse). Eventually, I arrived at a set of hyperparameters that does not lead to evident partial mode collapse during training. Very often, as training goes on, partial mode collapse disappears automatically and generated images look distinct from one another.

Another solution is to use Wasserstein loss which allows the discriminator to overcome the problem of vanishing gradient and prevents it from been stuck in a local minimum. As such, the generator can continue learning new modes to come up with better images.
